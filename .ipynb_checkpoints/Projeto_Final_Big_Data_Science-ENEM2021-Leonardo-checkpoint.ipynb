### Introdução:

"""O Exame Nacional do Ensino Médio (Enem) foi instituído em 1998, com o objetivo de avaliar o desempenho escolar dos estudantes ao término da educação básica. Em 2009, o exame aperfeiçoou sua metodologia e passou a ser utilizado como mecanismo de acesso à educação superior. Desde 2020, o participante pode escolher entre fazer o exame impresso ou o Enem Digital, com provas aplicadas em computadores, em locais de prova definidos pelo Inep.

As notas do Enem podem ser usadas para acesso ao Sistema de Seleção Unificada (Sisu) e ao Programa Universidade para Todos (ProUni). Elas também são aceitas em mais de 50 instituições de educação superior portuguesas. Além disso, os participantes do Enem podem pleitear financiamento estudantil em programas do governo, como o Fundo de Financiamento Estudantil (Fies). Os resultados do Enem possibilitam, ainda, o desenvolvimento de estudos e indicadores educacionais.

Qualquer pessoa que já concluiu o ensino médio ou está concluindo a etapa pode fazer o Enem para acesso à educação superior. Os participantes que ainda não concluíram o ensino médio podem participar como “treineiros” e seus resultados no exame servem somente para autoavaliação de conhecimentos.

A aplicação do Enem ocorre em dois dias. A Política de Acessibilidade e Inclusão do Inep garante atendimento especializado e tratamento pelo nome social, além de diversos recursos de acessibilidade. Há também uma aplicação para pessoas privadas de liberdade.

Os participantes fazem provas de quatro áreas de conhecimento: linguagens, códigos e suas tecnologias; ciências humanas e suas tecnologias; ciências da natureza e suas tecnologias; e matemática e suas tecnologias, que ao todo somam 180 questões objetivas. Os participantes também são avaliados por meio de uma redação, que exige o desenvolvimento de um texto dissertativo-argumentativo a partir de uma situação-problema."""

### Filtrar os dados sem os alunos que estão realizando a prova na modalidade de treineiro.

#### Também preciso tratar os dados de 2020 e 2021:

```python
dfB2 = dfB[dfB["IN_TREINEIRO"] == 0]

dfC2 = dfC[dfC["IN_TREINEIRO"] == 0]
```

```python
display(df2)
```

```python
print(f"O número de inscritos em 2019 foi de {len(df)}.")

print(f"O número de treineiros em 2020 foi de {len(dfB)}, o que representa um aumento de ~ 11,6 % em relação ao ano anterior.")

print(f"O número de treineiros em 2021 foi de {len(dfC)}, o que representa uma diminuição de ~ 40,4 % em relação ao ano anterior.")
```

```python
print(f"O número de treineiros em 2019 foi de {len(df)-len(df2)}.")

print(f"O número de treineiros em 2020 foi de {len(dfB)-len(dfB2)}.")

print(f"O número de treineiros em 2021 foi de {len(dfC)-len(dfC2)}.")
```

### Filtrar as notas somente com os alunos que estiveram presentes em todos os dias.

```python
start = time.time()

dftempa = df2[df2["TP_PRESENCA_CN"] == 1]
df3 = dftempa[dftempa["TP_PRESENCA_CH"] == 1]
dftempa = df3[df3["TP_PRESENCA_LC"] == 1]
df3 = dftempa[dftempa["TP_PRESENCA_MT"] == 1]

end = time.time()
print(f'Tempo para filtrar os dados: {end - start:2} segundos.')
```

#### Também preciso tratar 2020 e 2021:

```python
dftempa = dfB2[dfB2["TP_PRESENCA_CN"] == 1]
dfB3 = dftempa[dftempa["TP_PRESENCA_CH"] == 1]
dftempa = dfB3[dfB3["TP_PRESENCA_LC"] == 1]
dfB3 = dftempa[dftempa["TP_PRESENCA_MT"] == 1]
```

```python
dftempa = dfC2[dfC2["TP_PRESENCA_CN"] == 1]
dfC3 = dftempa[dftempa["TP_PRESENCA_CH"] == 1]
dftempa = dfC3[dfC3["TP_PRESENCA_LC"] == 1]
dfC3 = dftempa[dftempa["TP_PRESENCA_MT"] == 1]
```

```python
display(df3)
```

```python
#Plotando

dfpresentes = pd.DataFrame()
dfpresentes = pd.concat([df3[["NU_INSCRICAO", "NU_ANO"]], dfB3[["NU_INSCRICAO", "NU_ANO"]], dfC3[["NU_INSCRICAO", "NU_ANO"]]])

#display(dfpresentes)

dftempa = dfpresentes["NU_ANO"].value_counts()

plt.figure(figsize=(10,6))

plt.title("Número de alunos presentes em todos os dias da prova do ENEM, por ano:", fontdict=None, loc='center', fontweight=800)

x = dftempa.index
y = dftempa.values
for i in range(3):
    plt.annotate(dftempa.values[i], (dftempa.index[i], dftempa.values[i]+40000), ha="center")
    if i == 0:
        pass
    else:
        percent = (dftempa.values[i]/dftempa.values[i-1]) - 1
        plt.annotate((f"{percent:.2%}"), (dftempa.index[i],dftempa.values[i]-140000), ha="center")
    
plt.tick_params(left = False, right = False , labelleft = False , bottom = False)
plt.bar(x, y, color='#84ba5b')
plt.plot(x,y, color='r')
plt.xticks(x)
plt.show()
```

```python
print(f"O número de alunos que estiveram presentes em todos os dias de prova do ENEM em 2019 foi de {len(df3)}.")

print(f"O número de alunos que estiveram presentes em todos os dias de prova do ENEM em 2020 foi de {len(dfB3)}.")

print(f"O número de alunos que estiveram presentes em todos os dias de prova do ENEM em 2021 foi de {len(dfC3)}.")
```

### Criando uma nova variável da média geral das notas dos alunos.

```python
dict = {'ano': 2019, 'internet': 3954733}
dftempa = pd.DataFrame([dict])

dftempa = dftempa.append({'ano':2020, 'internet':4634811}, ignore_index=True)
dftempa = dftempa.append({'ano':2021, 'internet':3037632}, ignore_index=True)
```

```python
#Plotando

plt.figure(figsize=(10,6))

plt.suptitle("Número de alunos que possuem acesso à internet em casa", y=0.96, fontsize=16, fontweight=800)

plt.title("(e qual porcentagem, do total de inscritos naquele ano, isso representa).", fontdict=None, loc='center', fontweight=600)

dfinternet = dftempa.groupby("ano")["internet"].sum()

#display(dfinternet)

x = dfinternet.index
y = dfinternet.values
for i in range(3):
    plt.annotate(dfinternet.values[i], (dfinternet.index[i], dfinternet.values[i]+40000), ha="center")
    if i == 0:
        percent = dfinternet.values[i]/len(df)
        plt.annotate((f"{percent:.2%}"), (dfinternet.index[i], dfinternet.values[i]-180000), ha="center")
    elif i == 1:
        percent = dfinternet.values[i]/len(dfB)
        plt.annotate((f"{percent:.2%}"), (dfinternet.index[i], dfinternet.values[i]-180000), ha="center")
    else:
        percent = dfinternet.values[i]/len(dfC)
        plt.annotate((f"{percent:.2%}"), (dfinternet.index[i], dfinternet.values[i]-180000), ha="center")

    
plt.tick_params(left = False, right = False , labelleft = False , bottom = False)
plt.bar(x, y, color='orange')
plt.xticks(x)
plt.show()
```

### Observações e Conclusões

* O número de inscritos no ENEM aumentou um pouco em 2020 (aproximadamente 11,6% de aumento, comparado a 2019), e diminuiu drasticamente em 2021 (aproximadamente 40% de diminuição, comparado a 2020).

* Porém, o número de pessoas que estiveram presentes em todos os dias de prova do ENEM, em seus respectivos anos, diminuiu bastante em 2020 (aproximadamente 29% de diminuição, comparado a 2019), e diminuiu consideravelmente em 2021 (aproximadamente 16,5% de diminuição, comparado a 2020).

* A maioria dos alunos tinha, em 2019, a renda familiar entre 998,01 e 1.497,00 reais (C). Em 2021, a maioria dos alunos tinha a renda de até 1100,00 reais (B - uma categoria abaixo). Ou seja, a maior parte dos alunos inscritos no ENEM em 2021 estava uma categoria de renda abaixo quando comparado com o ano de 2019.

* E a maioria dos alunos que se inscreveram na prova possuem acesso à internet, número que vem crescendo. Em 2019, 77,6% do total dos alunos inscritos no ENEM possuiam acesso à internet. Já em 2021, 89,6% do total dos alunos inscritos no ENEM possuiam acesso à internet.

Obviamente estas análises demonstram o "comportamento" dos dados, mas não dizem respeito à causa desses comportamentos. Por exemplo, não podemos afirmar que a diminuição no número de inscritos na prova do ENEM durante o período é uma consequência somente da pandemia. Porém, são informações que evidenciam a necessidade do desenvolvimento de experimentos / pesquisas que consigam demonstrar a origem desse comportamento para que o órgão público possa tomar os devidos direcionamentos.
